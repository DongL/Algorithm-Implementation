## Implementation of Bubble Sort

```python
def bubble_sort(A):
    for i in range(len(A) - 1):
        for j in range(len(A) - 1, i, -1):
            if A[j] < A[j - 1]:
                A[j], A[j - 1] = A[j - 1], A[j]
```

### Self check

```python
A = [5,25,6,11,67,1,7,2,6,88,4,2]
bubble_sort(A)
print(A)
```

## Running time

```python
import numpy as np
import time
```

```python
# Load data
numbers = np.loadtxt('input-programming-1.txt')
```

```python
def get_averag_running_time(arr):
    running_times = list()
    for i in range(3):
        start = time.time()
        bubble_sort(arr)
        end = time.time()
        delta = end - start
        running_times.append(delta)
    return np.mean(running_times)
```

```python
obs = dict()
for n in range(500, 10500, 500):
    obs[n] = get_averaged_running_time(numbers[:n])
```

## Visually plot the measurements 

```python
import matplotlib.pyplot as plt
plt.style.use('default')
plt.plot(list(obs.keys()), list(obs.values()), 'o-')
plt.xlabel('Sample size (n)')
plt.ylabel('Average running time (seconds)')
plt.grid()
plt.title('Measurements of average running time of Bubble Sort')
plt.gcf().set_size_inches(8, 6)
```
